```python
import os
import pandas as pd
import shutil
```

```python
seq_dir = '/home/ayon/Dataset/Varos/2021-08-17_SEQ1'
img_dir = '/home/ayon/Dataset/Varos/2021-08-17_SEQ1/vehicle0/cam0/A'
time_stamps = '/home/ayon/Dataset/Varos/2021-08-17_SEQ1/vehicle0/cam0/camM0_timestamps.csv'
output_dir = '/home/ayon/git/orbslam3_docker/Datasets/EuRoC/MH01/mav0/cam0/data'
```

```python
#read the timestamps

df = pd.read_csv(time_stamps)
print(df.shape)

df.head()
```

```python
#create a rename dictionary
img_list = os.listdir(img_dir)
rename_dict = {}
for index, row in df.iterrows():
   
    rename_dict[row['A']] = row['timestamp [ns]']
    
assert len(img_list)==len(rename_dict)
```

```python
rename_dict.keys()
```

```python
def copy_and_rename(src_file, dest_file):
    try:
        shutil.copy(src_file, dest_file)
        print(f"File '{src_file}' copied and renamed to '{dest_file}'.")
    except FileNotFoundError:
        print(f"Error: File '{src_file}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

#rename and copy the files
for img in img_list:
    # print(img[:-4])
    new_name = str(rename_dict[img[:-4]])+'.png'
    # print(new_name)
    copy_and_rename(os.path.join(img_dir,img), os.path.join(output_dir, new_name))
    
```

```python
assert len(os.listdir(output_dir))==len(os.listdir(img_dir))
```
